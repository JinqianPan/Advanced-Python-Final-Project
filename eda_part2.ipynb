```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sqf_data.csv')
df.info()
```

```python
df.columns
```

```python
col1 =  ['additional_associating', 'additional_direction', 'additional_report',
       'additional_evasive', 'additional_highcrime',
       'additional_investigation', 'additional_proximity', 'additional_time',
       'additional_sights', 'additional_other', 'stopped_bulge',
       'stopped_object', 'stopped_casing', 'stopped_clothing', 'stopped_desc',
       'stopped_drugs', 'stopped_furtive', 'stopped_lookout',
       'stopped_violent', 'stopped_other']

tfdf = df[col1]
tfdf.head()
```

```python
count = pd.DataFrame(columns=col1)

for i in col1:
    per = tfdf[i].value_counts(normalize=True)[True]
    count.loc[0,i] = per
count
```

```python
plt.figure(figsize=(30,10))
plt.bar(range(len(col1)), count.iloc[0], tick_label =  col1)
plt.xticks(rotation = 45)
plt.xlabel('reasons')
plt.ylabel('posiblility')
plt.title('Posibility of stop by different reasons')
m = max(count.iloc[0])
plt.annotate(m, xy=(5,m))
plt.show()
```
